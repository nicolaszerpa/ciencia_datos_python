##***EJERCICIO 1***

Simular el lanzamiento de un dado eqilibrado de 6 caras.


*   Mostrar la probabilida aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
import random
import matplotlib.pyplot as plt

# Simulación del lanzamiento de un dado
dado = [random.randint(1, 6) for _ in range(1000)]

# Probabilidad aproximada de obtener los números 2, 3 y 5
prob_2 = dado.count(2) / len(dado)
prob_3 = dado.count(3) / len(dado)
prob_5 = dado.count(5) / len(dado)

print(f"Probabilidad de obtener el número 2: {prob_2:.2f}")
print(f"Probabilidad de obtener el número 3: {prob_3:.2f}")
print(f"Probabilidad de obtener el número 5: {prob_5:.2f}")

# Histograma de las probabilidades
plt.hist(dado, bins=6, density=True, alpha=0.7, color='g')
plt.xlabel('Número obtenido')
plt.ylabel('Probabilidad')
plt.title('Histograma del lanzamiento del dado')
plt.show()
```
