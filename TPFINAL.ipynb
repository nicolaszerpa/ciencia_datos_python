 Los valores atípicos son aquellos que están por debajo de Q1 - 1.5 * IQR o por encima de Q3 + 1.5 * IQR

```python
# Data Conversion
df["growth_percentage"] = df["growth_percentage"].str.replace("%", "").astype(float)
```

```python
def detect_outliers_iqr(data, column):
    q1 = data[column].quantile(0.25)
    q3 = data[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers
```

```python
# Llamamos a la función para la columna growth percentage
outliers = detect_outliers_iqr(df, "growth_percentage")

# Imprimimos la cantidad de valores atípicos detectados
print("Número de valores atípicos detectados:", len(outliers))
```

```python
# Hacer un histograma de growth_percentage
plt.figure(figsize=(10, 6))
plt.hist(df["growth_percentage"], bins=40)
plt.axvline(np.mean(df["growth_percentage"]), color="red", label="Media")
plt.axvline(np.median(df["growth_percentage"]), color="green", label="Mediana")
plt.xlabel("Porcentaje de crecimiento")
plt.ylabel("Frecuencia")
plt.title("Histograma de porcentaje de crecimiento")
plt.legend()
plt.show()

# Calcular estadísticas descriptivas de growth_percentage
mean = np.mean(df["growth_percentage"])
median = np.median(df["growth_percentage"])
q1 = np.percentile(df["growth_percentage"], 25)
q3 = np.percentile(df["growth_percentage"], 75)
iqr = q3 - q1
std = np.std(df["growth_percentage"])
cv = std / mean
```

```python
# Hacer un diagrama de caja de growth_percentage
plt.figure(figsize=(10, 6))
sns.boxplot(df["growth_percentage"], color="orange", showfliers=True)
plt.xlabel("Porcentaje de crecimiento")
plt.title("Diagrama de caja de porcentaje de crecimiento")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df["growth_percentage"], shade=True)
plt.title("Density plot of growth_percentage")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.ecdfplot(data=df)
plt.title('Quantile plot of growth_percentage')
plt.show()

# Imprimir los resultados
print(f"La media de growth_percentage es {round(mean, 2)}%")
print(f"La mediana de growth_percentage es {round(median, 2)}%")
print(f"El rango intercuartílico de growth_percentage es {round(iqr, 2)}%")
print(f"La desviación estándar de growth_percentage es {round(std, 2)}%")
print(f"El coeficiente de variación de growth_percentage es {round(cv, 2)}")
```

Una posible conclusión con los datos obtenidos del análisis univariado es:

La columna growth_percentage tiene una distribución asimétrica a la derecha, con una media de 105.84% y una mediana de 45%. Esto significa que hay algunas empresas que tienen un porcentaje de crecimiento muy alto, pero la mayoría tiene un porcentaje de crecimiento moderado o bajo.

La columna growth_percentage tiene muchos valores atípicos, que son los que se alejan mucho de la media o de los cuartiles. Estos valores pueden indicar que hay algunas empresas que tienen un rendimiento excepcional o que hay algún error en los datos.

La columna growth_percentage tiene una muy alta variabilidad, con una desviación estándar de 149.09% y un coeficiente de variación de 1.41. Esto significa que hay mucha dispersión en los datos y que el porcentaje de crecimiento puede variar mucho de una empresa a otra.

La columna growth_percentage tiene una forma unimodal(Por que tiene un solo pico en la media), con un solo pico en el gráfico de densidad. Esto puede indicar que hay un solo grupo de empresas con características similares, o que no hay ningún factor que influya en el porcentaje de crecimiento.

3-valuation y estimated_revenues: tienen una correlación positiva de 0.81, lo que significa que a mayor valoración de la empresa, mayor ingreso estimado.
